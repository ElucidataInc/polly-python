# Query for multiple sources and datatypes

Query structure for multiple source: 
-> query = """SELECT * FROM repo_name.source_name_in_schema.datasets 
        WHERE CONTAINS(curated_disease, 'Multiple Myeloma') """
        
Query structure for multiple source:
-> query = """SELECT * FROM repo_name.source_name_in_schema.datatype_name_in_schema.datasets 
        WHERE CONTAINS(curated_disease, 'Multiple Myeloma') """

```python
# ----------- QUERY EXAMPLE 1 -------------- # Query all datasets from GEO

query = """SELECT * FROM multiple_oa_2.geo.datasets"""
results=omixatlas.query_metadata(query)
results.head()
```

```python
# ----------- QUERY EXAMPLE 2 -------------- # Query all datasets from CBioPortal

query = """SELECT * FROM multiple_oa_2.cbioportal.datasets"""
results=omixatlas.query_metadata(query)
results.head()
```
